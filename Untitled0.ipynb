```python
!pip install yfinance
import yfinance as yf

data = yf.download("EURUSD=X", period="24mo", interval="1d")
data.head()


import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(data['Close'])
plt.title("EUR/USD - Prix de clture")
plt.xlabel("Date")
plt.ylabel("Prix")
plt.grid(True)
plt.show()


# Calcul des moyennes mobiles
data["SMA20"] = data["Close"].rolling(window=20).mean()
data["SMA50"] = data["Close"].rolling(window=50).mean()

# Affichage
plt.figure(figsize=(12,5))
plt.plot(data["Close"], label="Close Price")
plt.plot(data["SMA20"], label="SMA 20")
plt.plot(data["SMA50"], label="SMA 50")
plt.title("EUR/USD avec Moyennes Mobiles 20 & 50")
plt.xlabel("Date")
plt.ylabel("Prix")
plt.legend()
plt.grid(True)
plt.show()

```

```python
# Cration d'une colonne "Signal"
data["Signal"] = 0

# Achat quand SMA20 > SMA50
data["Signal"][data["SMA20"] > data["SMA50"]] = 1

# Vente quand SMA20 < SMA50
data["Signal"][data["SMA20"] < data["SMA50"]] = -1

data.tail(10)
```
